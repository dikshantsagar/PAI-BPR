```
from keras.models import Sequential
from keras.layers import Dense, Conv2D, MaxPool2D , Flatten
from keras import optimizers
import pickle
import numpy as np 
from keras.models import load_model
```

```
model = Sequential()
```

```
model.add(Conv2D(11, kernel_size=3, activation='relu', input_shape=(100,100,3),strides=4))
model.add(MaxPool2D(pool_size=2,strides=2))
model.add(Conv2D(5, kernel_size=3, activation='relu',strides=1))
model.add(MaxPool2D(pool_size=2,strides=2))
model.add(Conv2D(3, kernel_size=2, activation='relu',strides=1))
model.add(Conv2D(3, kernel_size=2, activation='relu',strides=1))
model.add(Conv2D(3, kernel_size=2, activation='relu',strides=1))
model.add(MaxPool2D(pool_size=2,strides=2))
model.add(Flatten())
model.add(Dense(4096, activation='relu'))
model.add(Dense(4096, activation='relu'))
model.add(Dense(2040,activation = 'relu',name='attribute'))
model.add(Dense(1000, activation='sigmoid'))        # Sigmoid cause output has 0 or 1 ans values in array (multiclass 1 hot encoded)
```

```
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])    #binary crossentropy cause target has only two values 0 and 1.
```

```
for i in range(1,7):
  t=i
  with open('/content/drive/My Drive/fashion/xtrain_'+str(t)+'.obj','rb') as f:    # loading all batches
    xt=pickle.load(f)
  xt=np.array(xt)
  
  with open('/content/drive/My Drive/fashion/ytrain_'+str(t)+'.obj','rb') as f:
    yt=pickle.load(f)
  yt=np.array(yt)

  xt=xt/255.  #normalizing the images
  model.fit(xt, yt, batch_size=32, epochs=5)
```
